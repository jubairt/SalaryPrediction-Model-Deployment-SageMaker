```python
import boto3
import json

# üîπ Replace with your endpoint name
ENDPOINT_NAME = "salary-endpoint"   # ‚Üê your actual endpoint name

# Create SageMaker runtime client
runtime = boto3.client(
    "sagemaker-runtime",
    region_name="us-east-1"
)

# üîπ Input data (must match training feature)
# YearsExperience ‚Üí Linear Regression input
payload = {
    "YearsExperience": [1.5, 3, 5, 8]
}
```

```python
# Invoke endpoint
response = runtime.invoke_endpoint(
    EndpointName=ENDPOINT_NAME,
    ContentType="application/json",
    Body=json.dumps(payload)
)

# Read response
result = response["Body"].read().decode("utf-8")
prediction = json.loads(result)
```
